```python
scores = [85, 92, 78, 88, 64]
students = ['철수', '영희', '길동', '민수', '지수']
for i in range(len(scores)):
    print(f'이름={students[i]} 점수={scores[i]}')
```

```python
total = 0
sum_txt = "총합="
for idx, score in enumerate(scores):
    #print(f"idx={idx}, score={score}")
    total += score
    #print(f"{students[idx]}({score})")
    sum_txt +=f"{students[idx]}({score})"
    if idx < len(scores)-1:
        sum_txt += '+'
    print(f"sum_txt = {sum_txt}")
print(f"{sum_txt}={total}점")
```

```python
def 가변상자(*goods):
    print("상자에 담긴 물건 목록 : ")
    for good in goods:
        print(f"-{good}")
```

```python
def 자기소개(이름, 나이, *취미들):
    print(f"제 이름은 {이름}이고 나이는 {나이}살 입니다.")
    if 취미들:
        print("제 취미는 다음과 같습니다")
        for 취미 in 취미들:
            print(f"-{취미}")
    else:
        print("취미는 아직 없습니다.")
```

```python
cart = []
while True:
    print("=======장바구니 프로그램=======")
    print("1. 물건 추가")
    print("2. 물건 삭제")
    print("3. 장바구니 보기")
    print("4. 프로그램 종료")
    choice = input("원하는 번호를 입력하세요 : ")

    match choice:
        case '1':
            cart.append(input("추가할 물건을 입력하세요 : "))
            print(f"현재 장바구니에는 {len(cart)}개의 물건이 있습니다.")
        case '2':
            thing = input("삭제할 물건을 입력하세요 : ")
            if thing in cart:
                cart.remove(thing)
                print(f"{thing}이 장바구니에서 빠졌습니다.")
            else:
                print(f"{thing}(은/는) 장바구니에 없는 물건입니다.")
        case '3':
            if not cart:
                print("현재 장바구니가 비었습니다.")
            else:
                print(f"장바구니 안에는 {cart}가 들어 있습니다!")
        case '4':
            break
        case _:
            print("잘못된 입력입니다 1~4번중에서 선택해주세요.")
```

```python
class Car:
    def __init__(self, color, brand, speed):
        self.color = color
        self.brand = brand
        self.speed = speed

    def drive(self):
        print(f"{self.brand}자동차가 {self.speed}km/h 속도로 달립니다")

    def stop(self):
        print("자동차가 멈췄습니다")
```

```python
class Product:
    def __init__(self, name, price, stock, intro):
        self.name = name
        self.price = price
        self.stock = stock
        self.intro = intro
    def show(self):
        print(f"상품명 : {self.name}", end="")
        print(f" 가격 : {self.price}원", end="")
        print(f" 재고 : {self.stock}개")
    def __str__(self):
        return f"{self.intro} 가격 : {self.price}\\"
    def discount(self, dis):
        print(f"이제 {dis}%의 할인이 적용됩니다")
        self.price *= (100-dis)/100
    def buy_product(self, qty):
        if qty <= self.stock:
            self.stock -= qty
            print(f"{qty}개 {qty*self.price}원 구매완료!! 남은 재고 {self.stock}개")
        else:
            print("재고가 부족합니다")
```

```python
note = Product("노트북", 1000000, 5, "삼성 노트북입니다")
note.show()
print(note)
```

```python
com = Product("컴퓨터", 1000000, 5, "-삼성 컴퓨터입니다-")
com.show()
print(com)
com.buy_product(1)
com.discount(30)
com.buy_product(3)
com.buy_product(5)
```
