```python
import requests

from googletrans import Translator

def fetch_joke_from_api(category='Programming'):
    url = f"https://v2.jokeapi.dev/joke/{category}?lang=en"
    Response = requests.get(url)
    data = Response.json()
    if data.get("type") == "single":
        return data.get("joke")
    else:
        return f'{data.get("setup")} - {data.get("delivery")}'

def translate_joke_to_korean(joke, lang="ko"):
    translator = Translator()
    translated_text = translator.translate(joke, dest=lang).text
    return translated_text

def translate_joke_to_de(joke, lang="ja"):
    translator = Translator()
    translated_text = translator.translate(joke, dest=lang).text
    return translated_text

joke = fetch_joke_from_api()
print(f"영어유머 : {joke}")
k_joke = translate_joke_to_korean(joke)
print(f"한글번역 : {k_joke}")
j_joke = translate_joke_to_de(joke)
print(f"일본어 : {j_joke}")

```
